Reduce fractions to lowest terms

```python
numerator = 12
denominator = 24
factor = 1

# Find greatest common factor
for test_factor in range(1,denominator+1):
    if numerator%test_factor==0 and denominator%test_factor==0:
        factor = test_factor

# Divide out greatest common factor
n = int(numerator/factor)
d = int(denominator/factor)

print("original: ", numerator, "/", denominator)
print("reduced: ", n, "/", d)
```

Divide out any perfect square factors. For example:<br>
 $\sqrt{12}$ = $\sqrt{4 * 3}$ = 2$\sqrt{3}$<br>
Because 4 is a square number, the square root of 4 is now outside the radical.

```python
import math

# number to factor
n = 12

# This variable will change
max_factor = 1

# The key ingredient
upper_limit = math.floor(math.sqrt(n)) + 1

# Find square factors
for maybe_factor in range(1,upper_limit):
    if n % (maybe_factor**2) == 0:
        max_factor = maybe_factor

# Results so far
print("n = ", n)
print("Square rooted factor = ", max_factor)
print("Square factor = ", max_factor**2)
print("integer: ", n/(max_factor**2))
```

#solving systems

Simplest way to solve a system of equations. Set each equation equal to zero

```python
from sympy import *

x,y = symbols('x y')

# Equations set equal to zero
first = 2*x + y - 1
second = x - 2*y + 7

# The solution
print(linsolve([first, second], (x, y)))
```

Nicer looking output

```python
from sympy import *

x,y = symbols('x y')

first = 2*x + y - 1
second = x - 2*y + 7


# parse finite set answer as coordinate pair
solution = linsolve([first, second], (x, y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]

# Print a coordinate pair
print("(", x_solution, ",", y_solution, ")")
```

Solve and graph. Factor to change from "`0 = `" to "`y = `" preparing to graph

```python
from sympy import *
from sympy.plotting import plot
from sympy import sqrt

var('x y')

# First equation set equal to zero, ready to solve
first = -x**2 - y + 10
#first = sqrt(x) - y

#Second equation set equal to zero, ready to solve
second = 2*x**2 - 2*y - 4
#second = -x + 5 - y

# Solve - can be linear or nonlinear equations
solution = nonlinsolve([first, second], (x, y))
for a in range(len(solution.args)):
    x_solution = solution.args[a][0]
    y_solution = solution.args[a][1]
    print("Solution = (", x_solution, ",", y_solution, ")")

# Sympy syntax for equation equal to zero, ready to factor
y_first = Eq(first,0)

# Sympy solve for y
y1 = solve(y_first,y)

# Same two steps for second equation
y_second = Eq(second,0)
y2 = solve(y_second,y)

# Show factored results
print("y = ", y1[0])
print("y = ", y2[0])

# Plot solution
x = symbols('x')
xmin = -10
xmax = 10
plot(y1[0], y2[0], (x,xmin,xmax))
```

#quadratics

Solve a quadratic equation

```python
# Coefficient values
a = 1
b = 5
c = 6

print("y = ", a, "x**2 + ", b, "x + ", c)

# Vertex
vx = -b/(2*a)
vy = a*(vx**2) + b*vx + c
print("Vertex: (",vx,",",vy,")")

# Roots
d = b**2 - 4*a*c
if d>=0:
    root_1 = (-b + math.sqrt(d))/(2*a)
    root_2 = (-b - math.sqrt(d))/(2*a)
    print("Roots: x = ", root_1, " and x = ", root_2)
else:
    print("No real roots")
```

Graph a quadratic function

```python
import matplotlib.pyplot as plt
import numpy as np
import math

a = 1
b = 5
c = 6

print("y = ", a, "x**2 + ", b, "x + ", c)

# Vertex
vx = -b/(2*a)
vy = a*(vx**2) + b*vx + c
print("Vertex: (",vx,",",vy,")")

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 10*(xmax-xmin)
x = np.linspace(xmin,xmax,points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

# Graph the parabola
y1 = a*x**2 + b*x + c
plt.plot(x, y1)

# Plot the vertex point
plt.plot([vx],[vy], 'ro')

# Find and plot the roots
d = b**2 - 4*a*c
if d>=0:
    root_1 = (-b + math.sqrt(d))/(2*a)
    root_2 = (-b - math.sqrt(d))/(2*a)
    plt.plot([root_1, root_2],[0,0], 'go')
    print("Roots: x = ", root_1, " and x = ", root_2)

plt.show()
```

Sliders to show how a, b, and c affect the graph

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import math

# All graphing happens in this function
def f(a,b,c):
    xmin = -10
    xmax = 10
    ymin = -10
    ymax = 10
    points = 10*(xmax-xmin)
    x = np.linspace(xmin,xmax,points)

    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'b') # x axis
    plt.plot([0,0],[ymin,ymax], 'b') # y axis

    # Parabola
    y1 = a*x**2 + b*x + c
    plt.plot(x, y1)

    # Vertex
    vx = -b/(2*a)
    vy = a*(vx**2) + b*vx + c
    plt.plot([vx],[vy], 'ro')

    # Roots
    d = b**2 - 4*a*c
    if d>=0:
        root_1 = (-b + math.sqrt(d))/(2*a)
        root_2 = (-b - math.sqrt(d))/(2*a)
        plt.plot([root_1, root_2],[0,0], 'go')

    # Set the equation as the title
    sa = str(a)
    sb = str(b)
    sc = str(c)
    h1 = "y = ", sa, "x**2 + ", sb, "x + ", sc
    h2 = ""
    for w in h1:
        h2 = h2 + w

    plt.title(h2)
    plt.show()

# Interactive creates sliders for a, b, and c values
interactive_plot = interactive(f, a=(1, 9), b=(-9,9), c=(-9,9))
interactive_plot
```

The algebra for any percent increase:<br>
New amount = p + p*r<br>
new amount = p * (1 + r)<br>

```python
p = 1000
r = .05
a = p*(1+r)
print("New amount = ", a)
```

Compounding is adding the interest to the principle and then calculating percent increase from this new amount<br>
Original amount = P<br>
New amount = P*(1+r)<br>
Compounded again = P*(1+r)*(1+r)<br>
And again = P*(1+r)*(1+r)*(1+r) = P(1+r)<sup>3</sup>

```python
P = 1000
r = .05
t = 3
annuity = p*(1+r)**t

print("Annuity = ", annuity)
```

When the compunding happens `n` times per year, here is the formula:<br>
A = P(1 +  $\frac{r}{n})^{nt}$

```python

P = 1000
r = .05
t = 3
n = 12

# use extra parentheses
annuity = p*(1+(r/n))**(n*t)

print("Annuity = ", annuity)
```

A good use of natural log: how long will it take for an investment to double in value?<br>
Start with the continuous growth formula:<br>
A = Pe<sup>rt</sup><br>
If the amount is double, then annuity is two times the principle:<br>
2P = Pe<sup>rt</sup><br>
Divide both sides by P and get this:<br>
2 = e<sup>rt</sup><br>
Because of the base <i>e</i>, take the natural log of both sides and get this:<br>
ln(2) = rt<br>
Then divide by r to solve for t.<br>
$\frac{ln(2)}{r}$ = t

```python
import math

# Rate of increase as a decimal
r = .09

# Natural log of 2 divided by rate
t = round(math.log(2)/r)

print("It will double in ", t, " years")
```

#multifunction calculator

```python

from sympy import symbols, Eq, solve
from fractions import Fraction
import math

class MultiFunctionCalculator:
    def __init__(self):
        pass

    def solve_proportion(self, a, b, c, x):
        """
        Solves the proportion a/b = c/x for x.
        """
        x_value = (b * c) / a
        return x_value

    def solve_for_x(self, equation):
        """
        Solves a linear equation for x.
        """
        x = symbols('x')
        eq = Eq(equation)
        solution = solve(eq, x)
        return solution

    def factor_square_root(self, n):
        """
        Factors the square root of n.
        """
        return math.sqrt(n)

    def decimal_to_fraction(self, decimal):
        """
        Converts a decimal to a fraction.
        """
        fraction = Fraction(decimal).limit_denominator()
        return fraction

    def decimal_to_percent(self, decimal):
        """
        Converts a decimal to a percent.
        """
        percent = decimal * 100
        return percent

    def fraction_to_decimal(self, numerator, denominator):
        """
        Converts a fraction to a decimal.
        """
        decimal = numerator / denominator
        return decimal

    def fraction_to_percent(self, numerator, denominator):
        """
        Converts a fraction to a percent.
        """
        decimal = self.fraction_to_decimal(numerator, denominator)
        percent = decimal * 100
        return percent

    def percent_to_decimal(self, percent):
        """
        Converts a percent to a decimal.
        """
        decimal = percent / 100
        return decimal

    def percent_to_fraction(self, percent):
        """
        Converts a percent to a fraction.
        """
        decimal = self.percent_to_decimal(percent)
        fraction = self.decimal_to_fraction(decimal)
        return fraction

# Usage example
calculator = MultiFunctionCalculator()

# Solving a proportion
print("Solve proportion 2/3 = 4/x:", calculator.solve_proportion(2, 3, 4, 'x'))

# Solving for x in equation 2x - 4 = 0
x = symbols('x')
print("Solve for x in 2x - 4 = 0:", calculator.solve_for_x(2*x - 4))

# Factoring square root
print("Factor square root of 16:", calculator.factor_square_root(16))

# Decimal to fraction and percent
print("Convert decimal 0.75 to fraction:", calculator.decimal_to_fraction(0.75))
print("Convert decimal 0.75 to percent:", calculator.decimal_to_percent(0.75))

# Fraction to decimal and percent
print("Convert fraction 3/4 to decimal:", calculator.fraction_to_decimal(3, 4))
print("Convert fraction 3/4 to percent:", calculator.fraction_to_percent(3, 4))

# Percent to decimal and fraction
print("Convert percent 75% to decimal:", calculator.percent_to_decimal(75))
print("Convert percent 75% to fraction:", calculator.percent_to_fraction(75))
```

#graphing calculator

```python

import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, Eq, solve
from sympy.parsing.sympy_parser import parse_expr

class GraphingCalculator:
    def __init__(self):
        self.x = symbols('x')

    def plot_function(self, functions, x_range=(-10, 10), shade=None):
        """
        Plots one or more functions.
        :param functions: List of function expressions as strings.
        :param x_range: Tuple of (xmin, xmax) to set the range of x-axis.
        :param shade: 'above' or 'below' to shade the area.
        """
        x_vals = np.linspace(x_range[0], x_range[1], 400)
        plt.figure()

        for func in functions:
            f = parse_expr(func)
            y_vals = [f.subs(self.x, x) for x in x_vals]
            plt.plot(x_vals, y_vals, label=f"${func}$")

            if shade:
                y_fill = np.maximum(y_vals, 0) if shade == 'above' else np.minimum(y_vals, 0)
                plt.fill_between(x_vals, y_vals, y_fill, where=(y_vals > 0 if shade == 'above' else y_vals < 0), alpha=0.2)

        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Function Plot')
        plt.axhline(0, color='black',linewidth=0.5)
        plt.axvline(0, color='black',linewidth=0.5)
        plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
        plt.legend()
        plt.show()

    def create_table(self, func, x_values):
        """
        Creates a table of (x,y) values.
        :param func: Function expression as a string.
        :param x_values: List of x values.
        :return: List of (x, y) tuples.
        """
        f = parse_expr(func)
        table = [(x, f.subs(self.x, x)) for x in x_values]
        return table

    def solve_and_graph_system(self, equations, x_range=(-10, 10)):
        """
        Solves and graphs a system of equations.
        :param equations: List of equations as strings.
        :param x_range: Tuple of (xmin, xmax) to set the range of x-axis.
        """
        eqs = [Eq(parse_expr(eq.split('=')[0]), parse_expr(eq.split('=')[1])) for eq in equations]
        solutions = solve(eqs, (self.x, symbols('y')))

        if solutions:
            x_vals = np.linspace(x_range[0], x_range[1], 400)
            plt.figure()

            for eq in equations:
                f = parse_expr(eq.split('=')[0])
                y_vals = [f.subs(self.x, x) for x in x_vals]
                plt.plot(x_vals, y_vals, label=f"${eq}$")

            plt.xlabel('x')
            plt.ylabel('y')
            plt.title('System of Equations')
            plt.axhline(0, color='black',linewidth=0.5)
            plt.axvline(0, color='black',linewidth=0.5)
            plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
            plt.legend()
            plt.show()

        return solutions

    def solve_quadratic(self, a, b, c):
        """
        Solves a quadratic equation ax^2 + bx + c = 0.
        :param a: Coefficient of x^2.
        :param b: Coefficient of x.
        :param c: Constant term.
        :return: Solutions of the quadratic equation.
        """
        quadratic_eq = Eq(a*self.x**2 + b*self.x + c, 0)
        solutions = solve(quadratic_eq, self.x)
        return solutions

# Usage example
calc = GraphingCalculator()

# Graph one or more functions
calc.plot_function(["x**2", "x**3"], x_range=(-10, 10))

# Create a table of (x,y) values
table = calc.create_table("x**2", [-2, -1, 0, 1, 2])
print("Table of (x,y) values for x^2:", table)

# Solve and graph a system of equations
solutions = calc.solve_and_graph_system(["x + y = 2", "x - y = 0"], x_range=(-10, 10))
print("Solutions for the system of equations:", solutions)

# Solve quadratic equation
quad_solutions = calc.solve_quadratic(1, -3, 2)
print("Solutions for quadratic equation x^2 - 3x + 2 = 0:", quad_solutions)
```

#three math games

```python

import tkinter as tk
import matplotlib.pyplot as plt
import numpy as np
import random

class ScatterPlotGame:
    def __init__(self, size=10):
        self.size = size
        self.points = self.generate_points()
        self.correct_answers = 0
        self.total_questions = len(self.points)

        self.root = tk.Tk()
        self.root.title("Scatter Plot Game")

        self.canvas = tk.Canvas(self.root, width=600, height=600)
        self.canvas.pack()

        self.entry_x = tk.Entry(self.root)
        self.entry_y = tk.Entry(self.root)
        self.entry_x.pack()
        self.entry_y.pack()

        self.submit_button = tk.Button(self.root, text="Submit", command=self.check_answer)
        self.submit_button.pack()

        self.label = tk.Label(self.root, text="")
        self.label.pack()

        self.plot_points()
        self.root.mainloop()

    def generate_points(self):
        points = []
        for _ in range(10):
            x = random.randint(-self.size, self.size)
            y = random.randint(-self.size, self.size)
            points.append((x, y))
        return points

    def plot_points(self
        plt.figure()
        for x, y in self.points:
            plt.scatter(x, y)
        plt.xlim(-self.size, self.size)
        plt.ylim(-self.size, self.size)
        plt.axhline(0, color='black',linewidth=0.5)
        plt.axvline(0, color='black',linewidth=0.5)
        plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
        plt.show()

    def check_answer(self):
        try:
            x = int(self.entry_x.get())
            y = int(self.entry_y.get())
            if (x, y) in self.points:
                self.correct_answers += 1
                self.label.config(text=f"Correct! {self.correct_answers}/{self.total_questions}")
            else:
                self.label.config(text=f"Wrong. {self.correct_answers}/{self.total_questions}")
        except ValueError:
            self.label.config(text="Please enter valid integers.")

        if self.correct_answers == self.total_questions:
            self.label.config(text="Game Over! You've identified all points.")

# Run the game
scatter_game = ScatterPlotGame(size=15)
```

```python

import math
import tkinter as tk
from sympy import symbols, Eq, solve, log

class FinancialCalculator:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Financial Calculator")

        self.create_widgets()
        self.root.mainloop()

    def create_widgets(self):
        # Create input fields and labels for each function
        tk.Label(self.root, text="Calculate Annuity").grid(row=0, column=0, padx=10, pady=10)
        tk.Label(self.root, text="Principal:").grid(row=1, column=0)
        self.annuity_principal = tk.Entry(self.root)
        self.annuity_principal.grid(row=1, column=1)
        tk.Label(self.root, text="Rate (annual):").grid(row=2, column=0)
        self.annuity_rate = tk.Entry(self.root)
        self.annuity_rate.grid(row=2, column=1)
        tk.Label(self.root, text="Time (years):").grid(row=3, column=0)
        self.annuity_time = tk.Entry(self.root)
        self.annuity_time.grid(row=3, column=1)
        self.annuity_type = tk.StringVar(value="monthly")
        tk.Radiobutton(self.root, text="Monthly", variable=self.annuity_type, value="monthly").grid(row=4, column=0)
        tk.Radiobutton(self.root, text="Continuous", variable=self.annuity_type, value="continuous").grid(row=4, column=1)
        tk.Button(self.root, text="Calculate Annuity", command=self.calculate_annuity).grid(row=5, column=0, columnspan=2)

        tk.Label(self.root, text="Calculate Monthly Mortgage Payment").grid(row=6, column=0, padx=10, pady=10)
        tk.Label(self.root, text="Loan Amount:").grid(row=7, column=0)
        self.mortgage_amount = tk.Entry(self.root)
        self.mortgage_amount.grid(row=7, column=1)
        tk.Label(self.root, text="Rate (annual):").grid(row=8, column=0)
        self.mortgage_rate = tk.Entry(self.root)
        self.mortgage_rate.grid(row=8, column=1)
        tk.Label(self.root, text="Time (years):").grid(row=9, column=0)
        self.mortgage_time = tk.Entry(self.root)
        self.mortgage_time.grid(row=9, column=1)
        tk.Button(self.root, text="Calculate Mortgage", command=self.calculate_mortgage).grid(row=10, column=0, columnspan=2)

        tk.Label(self.root, text="Estimate Retirement Investment Balance").grid(row=11, column=0, padx=10, pady=10)
        tk.Label(self.root, text="Initial Investment:").grid(row=12, column=0)
        self.retirement_principal = tk.Entry(self.root)
        self.retirement_principal.grid(row=12, column=1)
        tk.Label(self.root, text="Rate (annual):").grid(row=13, column=0)
        self.retirement_rate = tk.Entry(self.root)
        self.retirement_rate.grid(row=13, column=1)
        tk.Label(self.root, text="Monthly Contribution:").grid(row=14, column=0)
        self.retirement_contribution = tk.Entry(self.root)
        self.retirement_contribution.grid(row=14, column=1)
        tk.Label(self.root, text="Time (years):").grid(row=15, column=0)
        self.retirement_time = tk.Entry(self.root)
        self.retirement_time.grid(row=15, column=1)
        tk.Button(self.root, text="Estimate Retirement Balance", command=self.estimate_retirement_balance).grid(row=16, column=0, columnspan=2)

        tk.Label(self.root, text="Calculate Doubling Time").grid(row=17, column=0, padx=10, pady=10)
        tk.Label(self.root, text="Rate (annual):").grid(row=18, column=0)
        self.doubling_rate = tk.Entry(self.root)
        self.doubling_rate.grid(row=18, column=1)
        tk.Button(self.root, text="Calculate Doubling Time", command=self.calculate_doubling_time).grid(row=19, column=0, columnspan=2)

        tk.Label(self.root, text="Solve Logarithmic Equation").grid(row=20, column=0, padx=10, pady=10)
        tk.Label(self.root, text="Log Equation (e.g., log(x,2)=3):").grid(row=21, column=0)
        self.log_equation = tk.Entry(self.root)
        self.log_equation.grid(row=21, column=1)
        tk.Button(self.root, text="Solve Log Equation", command=self.solve_logarithmic_equation).grid(row=22, column=0, columnspan=2)

        tk.Label(self.root, text="Convert to/from Scientific Notation").grid(row=23, column=0, padx=10, pady=10)
        tk.Label(self.root, text="Number:").grid(row=24, column=0)
        self.scientific_number = tk.Entry(self.root)
        self.scientific_number.grid(row=24, column=1)
        tk.Button(self.root, text="Convert to Scientific Notation", command=self.convert_to_scientific).grid(row=25, column=0, columnspan=2)
        tk.Button(self.root, text="Convert from Scientific Notation", command=self.convert_from_scientific).grid(row=26, column=0, columnspan=2)

        self.result_label = tk.Label(self.root, text="")
        self.result_label.grid(row=27, column=0, columnspan=2)

    def calculate_annuity(self):
        principal = float(self.annuity_principal.get())
        rate = float(self.annuity_rate.get()) / 100
        time = float(self.annuity_time.get())
        if self.annuity_type.get() == "monthly":
            n = 12
            amount = principal * ((1 + rate/n)**(n*time))
        else:
            amount = principal * math.exp(rate*time)
        self.result_label.config(text=f"Annuity Value: ${amount:.2f}")

    def calculate_mortgage(self):
        amount = float(self.mortgage_amount.get())
        rate = float(self.mortgage_rate.get()) / 100 / 12
        time = int(self.mortgage_time.get()) * 12
        payment = amount * (rate * (1 + rate)**time) / ((1 + rate)**time - 1)
        self.result_label.config(text=f"Monthly Mortgage Payment: ${payment:.2f}")

    def estimate_retirement_balance(self):
        principal = float(self.retirement_principal.get())
        rate = float(self.retirement_rate.get()) / 100 / 12
        contribution = float(self.retirement_contribution.get())
        time = int(self.retirement_time.get()) * 12
        future_value = principal * (1 + rate)**time + contribution * (((1 + rate)**time - 1) / rate)
        self.result_label.config(text=f"Retirement Balance: ${future_value:.2f}")

    def calculate_doubling_time(self):
        rate = float(self.doubling_rate.get()) / 100
        doubling_time = math.log(2) / math.log(1 + rate)
        self.result_label.config(text=f"Doubling Time: {doubling_time:.2f} years")

    def solve_logarithmic_equation(self):
        equation = self.log_equation.get()
        x = symbols('x')
        if "log" in equation:
            base = int(equation.split('log(')[-1].split(')')[0].split(',')[1])
            value = int(equation.split('=')[-1])
            solved_value = base**value
        else:
            solved_value = solve(Eq(eval(equation.split('=')[0]), int(equation.split('=')[-1])), x)
        self.result_label.config(text=f"Solved Value: {solved_value}")

    def convert_to_scientific(self):
        number = float(self.scientific_number.get())
        scientific_notation = "{:.2e}".format(number)
        self.result_label.config(text=f"Scientific Notation: {scientific_notation}")

    def convert_from_scientific(self):
        number = self.scientific_number.get()
        if 'e' in number:
            real_number = float(number)
        else:
            real_number = number
        self.result_label.config(text=f"Real Number: {real_number}")

# Run the financial calculator
financial_calculator = FinancialCalculator()
```

```python

import tkinter as tk
from tkinter import filedialog
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlopen

class DataGraphExplorer:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Data Graph Explorer")
        self.dataframe = None
        self.column_names = []

        self.create_widgets()
        self.root.mainloop()

    def create_widgets(self):
        # Upload local file
        self.upload_button = tk.Button(self.root, text="Upload CSV", command=self.upload_csv)
        self.upload_button.pack(pady=10)

        # URL input
        tk.Label(self.root, text="Enter URL for CSV:").pack()
        self.url_entry = tk.Entry(self.root, width=50)
        self.url_entry.pack()
        self.url_button = tk.Button(self.root, text="Load CSV from URL", command=self.load_csv_from_url)
        self.url_button.pack(pady=10)

        # URL hardcoded (example)
        self.url_hardcoded_button = tk.Button(self.root, text="Load CSV from Hardcoded URL", command=self.load_csv_from_hardcoded_url)
        self.url_hardcoded_button.pack(pady=10)

        # Display headings and first two rows
        self.display_button = tk.Button(self.root, text="Display Headings and First Two Rows", command=self.display_headings_and_rows)
        self.display_button.pack(pady=10)

        # Plot data
        tk.Label(self.root, text="Select columns to plot:").pack()
        self.column1 = tk.StringVar(self.root)
        self.column2 = tk.StringVar(self.root)
        self.column1_menu = tk.OptionMenu(self.root, self.column1, *self.column_names)
        self.column1_menu.pack()
        self.column2_menu = tk.OptionMenu(self.root, self.column2, *self.column_names)
        self.column2_menu.pack()

        self.plot_type = tk.StringVar(self.root, value="scatter")
        tk.Radiobutton(self.root, text="Scatter Plot", variable=self.plot_type, value="scatter").pack()
        tk.Radiobutton(self.root, text="Line Graph", variable=self.plot_type, value="line").pack()

        self.plot_button = tk.Button(self.root, text="Plot Data", command=self.plot_data)
        self.plot_button.pack(pady=10)

        self.result_label = tk.Label(self.root, text="")
        self.result_label.pack(pady=10)

    def upload_csv(self):
        file_path = filedialog.askopenfilename(filetypes=[("CSV files", "*.csv")])
        if file_path:
            self.load_csv(file_path)

    def load_csv_from_url(self):
        url = self.url_entry.get()
        if url:
            self.load_csv(urlopen(url))

    def load_csv_from_hardcoded_url(self):
        url = "https://example.com/data.csv"  # Replace with your actual URL
        self.load_csv(urlopen(url))

    def load_csv(self, file):
        self.dataframe = pd.read_csv(file)
        self.column_names = self.dataframe.columns.tolist()
        self.update_column_menus()

    def update_column_menus(self):
        self.column1_menu['menu'].delete(0, 'end')
        self.column2_menu['menu'].delete(0, 'end')
        for col in self.column_names:
            self.column1_menu['menu'].add_command(label=col, command=tk._setit(self.column1, col))
            self.column2_menu['menu'].add_command(label=col, command=tk._setit(self.column2, col))

    def display_headings_and_rows(self):
        if self.dataframe is not None:
            headings = self.dataframe.columns.tolist()
            first_two_rows = self.dataframe.head(2)
            self.result_label.config(text=f"Headings: {headings}\nFirst two rows:\n{first_two_rows}")
        else:
            self.result_label.config(text="No data loaded.")

    def plot_data(self):
        if self.dataframe is not None:
            col1 = self.column1.get()
            col2 = self.column2.get()
            if col1 and col2:
                x = self.dataframe[col1].to_numpy()
                y = self.dataframe[col2].to_numpy()

                plt.figure()
                if self.plot_type.get() == "scatter":
                    plt.scatter(x, y)
                    plt.title(f"Scatter Plot of {col1} vs {col2}")
                else:
                    plt.plot(x, y)
                    plt.title(f"Line Graph of {col1} vs {col2}")
                plt.xlabel(col1)
                plt.ylabel(col2)
                plt.show()
            else:
                self.result_label.config(text="Please select both columns to plot.")
        else:
            self.result_label.config(text="No data loaded.")

# Run the data graph explorer
data_graph_explorer = DataGraphExplorer()
```
